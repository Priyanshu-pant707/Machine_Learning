### function to handle the missing value

```python
def handle_missing_values(df, strategy='auto'):
   
    
    # 1. Count missing values in each column
    print("Missing values per column:\n")
    print(df.isnull().sum())
    print("\n")
    
    if strategy == 'remove':
        # Remove rows with any missing values
        df_cleaned = df.dropna()
        print("Rows with missing values removed.")
        return df_cleaned
    
    # Automatic handling based on column type
    df_cleaned = df.copy()
    
    for col in df_cleaned.columns:
        if df_cleaned[col].isnull().sum() > 0:  # If there are missing values
            if df_cleaned[col].dtype in ['int64', 'float64']:  # Numerical column
                # Fill with mean safely
                df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mean())
                print(f"Filled missing values in numerical column '{col}' with mean.")
            else:  # Categorical column
                # Fill with mode safely
                mode_value = df_cleaned[col].mode()[0]
                df_cleaned[col] = df_cleaned[col].fillna(mode_value)
                print(f"Filled missing values in categorical column '{col}' with mode '{mode_value}'.")
    
    return df_cleaned
```

```python
df_cleaned = handle_missing_values(df)
df_cleaned
```
